```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
products= pd.read_csv(r'C:\Users\singh\OneDrive\Desktop\projects\Blinkit\blinkit_products.csv')
orders= pd.read_csv(r'C:\Users\singh\OneDrive\Desktop\projects\Blinkit\blinkit_orders.csv')
orders_items= pd.read_csv(r'C:\Users\singh\OneDrive\Desktop\projects\Blinkit\blinkit_order_items.csv')
```

```python
merge1 = pd.merge(orders_items, products, on= 'product_id', how='left')
```

```python
merge1.head()
```

```python
df = pd.merge(merge1, orders, on= 'order_id', how='left')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Set plot style
sns.set_style("whitegrid")
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["xtick.labelsize"] = 10
plt.rcParams["ytick.labelsize"] = 10
```

```python
plt.figure(figsize=(12,5))
```

```python


sns.countplot(data = df, y = 'category', order = df['category'].value_counts().index )
plt.title('Category Count')
plt.xlabel('Count')
plt.ylabel('Category Name')
plt.show()
```

### FINDING_1 = Top five ordered items are: (1) Dairy & Breakfast products. (2) Householdcare products (3) Pet Care. (4) Fruit & Vegetables (5) Snacks & Munchies


```python
# ---- 2. Brand-wise Product Count (Top 10 Brands) ----

top_brands = df["brand"].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x = top_brands.values, y = top_brands.index, palette = 'BuGn_r')
plt.title("Top 10 Brands by Product Count")
plt.xlabel("Number of Products")
plt.ylabel("Brand")
plt.show()
```

```python
#margine percentge distribution

plt.figure(figsize=(8,8))
sns.histplot(df['margin_percentage'], bins=10, kde= True, color= 'g')
plt.title('Margin Distribution')
plt.xlabel('Margin%')
plt.ylabel('Frequency')
plt.show()
```

### No well defined distribution pattern has been observed in case of margine_percentage. However, the hisplot dictates that the shop is making a margin percentage between 20% to 35% on maximum number of products.

```python
plt.figure(figsize=(8,8))
sns.histplot(df['unit_price'], bins=30, kde= True, color= 'g')
plt.title('Price Distribution')
plt.xlabel('unitprice')
plt.ylabel('Frequency')
plt.show()
```
